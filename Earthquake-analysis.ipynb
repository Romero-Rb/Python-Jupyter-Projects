#### Visualizing the average magnitude of the earthquakes each year

```python
dtimemag = dframe[['date','magnitudo']].copy()
```

```python
dframe_avg = dtimemag.groupby(dtimemag['date'].dt.year).mean()
```

```python
dframe_avg
```

```python
pt.scatter(dframe_avg['date'].dt.year, dframe_avg['magnitudo'], c = "red")
pt.xlabel("Years")
pt.ylabel("Average Magnitude")
pt.title("Average Magnitude of Earthquakes for Each Year")
pt.show()
```

#### Calculating the average deviation for the magnitudes of each year
Turning data into an array to calculate the average

```python
allmagavg = dframe_avg['magnitudo'].mean()
```

```python
avg_mags = dframe_avg['magnitudo'].to_numpy()
```

```python
avg_mags
```

```python
finavg_mags = ny.array([])
```

```python
for x in avg_mags:
    finavg_mags = ny.append(finavg_mags,abs(allmagavg - x))
```

```python
finavg_mags
```

```python
finavg_mags = finavg_mags.mean()
```
